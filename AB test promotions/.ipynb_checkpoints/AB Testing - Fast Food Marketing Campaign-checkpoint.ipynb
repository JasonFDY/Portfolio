# A/B Testing - Fast Food Marketing Campaign

### Scenario:
A fast-food chain plans to add a new item to its menu. However, they are still undecided between three possible marketing campaigns for promoting the new product. In order to determine which promotion has the greatest effect on sales, the new item is introduced at locations in several randomly selected markets. A different promotion is used at each location, and the weekly sales of the new item are recorded for the first four weeks.

### Goal:
Evaluate A/B testing results and decide which marketing strategy works the best.

### Columns:

- MarketID: unique identifier for market

- MarketSize: size of market area by sales

- LocationID: unique identifier for store location

- AgeOfStore: age of store in years

- Promotion: one of three promotions that were tested

- week: one of four weeks when the promotions were run

- SalesInThousands: sales amount for a specific LocationID, Promotion, and week

### Hypothesis Testing

The null hypothesis is that there is no difference between the promotions. This will be tested using all the combined data for all market ids as well as the split data where market id 3 is separated. The reasoning for this is that the distribution for the data becomes a lot more skewed with the inclusion of market id 3 as it would be wise to test the impact of this.

As there are three groups being compared, an ANOVA test will be used with a significance threshold of 0.05. If the result is significant, the null hypothesis will be rejected and a tukey test will be used to determine between which promotions the null hypothesis has been rejected and decide which promotion is the best.

#### All Data

```python
promotion_1_all = df[df['Promotion']==1]['SalesInThousands']
promotion_2_all = df[df['Promotion']==2]['SalesInThousands']
promotion_3_all = df[df['Promotion']==3]['SalesInThousands']

fstat, pval = f_oneway(promotion_1_all, promotion_2_all, promotion_3_all)
print(pval)
```

The result of this ANOVA test is significant and the null hypothesis has been rejected.

```python
tukey_results = pairwise_tukeyhsd(df.SalesInThousands, df.Promotion, 0.05)
print(tukey_results)
```

The result of this tukey test rejects the null hypotheses between promotions 1 and 2 and promotions 2 and 3. It doesn't reject the null hypothesis between promotions 1 and 3. This would suggest that promotion 2 is significantly different to the other promotions but that promotions 1 and 3 are not. For these tests, the assumption that the data is normally distributed has been violated which means that the results may be incorrect.

#### Market ID 3

```python
id_3 = df[df['MarketID']==3]

promotion_1_id_3 = id_3[id_3['Promotion']==1]['SalesInThousands']
promotion_2_id_3 = id_3[id_3['Promotion']==2]['SalesInThousands']
promotion_3_id_3 = id_3[id_3['Promotion']==3]['SalesInThousands']

fstat, pval = f_oneway(promotion_1_id_3, promotion_2_id_3, promotion_3_id_3)
print(pval)
```

The result of this ANOVA test is significant and the null hypothesis has been rejected.

```python
tukey_results = pairwise_tukeyhsd(id_3.SalesInThousands, id_3.Promotion, 0.05)
print(tukey_results)
```

The result of this tukey test rejects the null hypotheses between all 3 of the promotions. This would suggest that all of the promotions are signifcantly different from one another and that promotion 1 is the best, promotion 3 is second and promotion 2 is last.

#### Market IDs 1,2,4,5,6,7,8,9,10

```python
not_3 = df[df['MarketID'].isin([1,2,4,5,6,7,8,9,10])]

promotion_1_not_3 = not_3[not_3['Promotion']==1]['SalesInThousands']
promotion_2_not_3 = not_3[not_3['Promotion']==2]['SalesInThousands']
promotion_3_not_3 = not_3[not_3['Promotion']==3]['SalesInThousands']

fstat, pval = f_oneway(promotion_1_not_3, promotion_2_not_3, promotion_3_not_3)
print(pval)
```

The result of this ANOVA test is significant and the null hypothesis has been rejected.

```python
tukey_results = pairwise_tukeyhsd(not_3.SalesInThousands, not_3.Promotion, 0.05)
print(tukey_results)
```

The result of this tukey test rejects the null hypotheses between all 3 of the promotions. This would suggest that all of the promotions are signifcantly different from one another and that promotion 1 is the best, promotion 3 is second and promotion 2 is last.

### Conclusion

The fast food company should focus on promotion 1 for all of the markets.
